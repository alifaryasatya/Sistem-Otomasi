```python
import numpy as np
import matplotlib.pyplot as plt

# Waktu dari 0 sampai 5 detik
t = np.linspace(0, 5, 500)

# Daftar nilai s
s_values = [0, 3, -1, 1j, 3 + 1j]
labels = ['s=0', 's=3', 's=-1', 's=j', 's=3+j']

# Plot
plt.figure(figsize=(12, 6))

for s, label in zip(s_values, labels):
    f_t = np.exp(s * t)
    plt.plot(t, np.real(f_t), label=f'Re({label})')
    if np.imag(s) != 0:
        plt.plot(t, np.imag(f_t), '--', label=f'Im({label})')

plt.title('Plot f(t) = e^(s¬∑t) untuk berbagai nilai s')
plt.xlabel('t (waktu)')
plt.ylabel('f(t)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
